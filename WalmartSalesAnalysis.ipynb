## Insights and Understanding the Dataset
1) a) Total weekly sales of each walmart store on Normal week Total weekly sales of each walmart store on Holiday weeks
   b) Total weekly sales of each walmart store on Holiday weeks 
2) a) Avg weekly sales of each walmart store
   b) Avg weekly sales of each walmart store on Holiday weeks.


```python
#1) a)Total weekly sales of each walmart store
Tot_week_sales_Normal = dframe[dframe['Holiday_Flag']==0].groupby('Store')['Weekly_Sales'].sum().astype(int)
print('Total Weekly Sales on a normal day for each store:', Tot_week_sales_Normal)
```

```python
#b) Total weekly Sales of each walmart store on Holiday weeks 
Tot_week_sales_Holiday = dframe[dframe['Holiday_Flag']==1].groupby('Store')['Weekly_Sales'].sum().astype(int)

print('Total weekly sales for each store on holiday week:',Tot_week_sales_Holiday)
```

```python
#2 a) Average weekly sale of each walmart store in a normal week 
Avg_NormalWeek_sale = dframe[dframe['Holiday_Flag']==0].groupby('Store')['Weekly_Sales'].mean().astype(int)
print('Average weekly sale on a normal day for each Walmart store:', Avg_NormalWeek_sale)
```

```python
#b) Average weekly sales of each wlamrt store in a holiday week
Avg_HolidayWeek_sale = dframe[dframe['Holiday_Flag']==1].groupby('Store')['Weekly_Sales'].mean().astype(int)

print('Average Sales in Holiday Week is:',Avg_HolidayWeek_sale)
```

 ## Visualization and Reports based on the analysis above on whether holiday weeks have an impact on the sales compared to normal week

```python
# Bar plot for comparing total weekly sales of each walmart store on Holiday/Normal week 

Tot_week_sales_Normal = dframe[dframe['Holiday_Flag']==0].groupby('Store')['Weekly_Sales'].sum().astype(int)
Tot_week_sales_Holiday = dframe[dframe['Holiday_Flag'] == 1].groupby('Store')['Weekly_Sales'].sum().astype(int)
Avg_NormalWeek_sale = dframe[dframe['Holiday_Flag']==0].groupby('Store')['Weekly_Sales'].mean().astype(int)
Avg_HolidayWeek_sale = dframe[dframe['Holiday_Flag'] == 1].groupby('Store')['Weekly_Sales'].mean().astype(int)

stores = Tot_week_sales_Normal.index #storing id 
plt.figure(figsize=(10,7))
bar_width = 0.6 
x = np.arange(len(stores))

plt.bar(x - bar_width / 2, Tot_week_sales_Normal, width=bar_width, label='Total Sales on a normal week ',color='cyan')
plt.bar(x + bar_width / 2, Tot_week_sales_Holiday, width=bar_width, label='Total Sales from just holiday week',color='Red')

plt.xlabel('Stores')
plt.ylabel('Total Weekly Sales')
plt.title('Total Weekly Sales: Holiday vs Non-Holiday Weeks')
plt.xticks(x, stores, rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Bar plot for comparing Average weekly sales of each walmart store on Holiday/Normal week 
Tot_week_sales_Normal = dframe[dframe['Holiday_Flag']==0].groupby('Store')['Weekly_Sales'].sum().astype(int)
Tot_week_sales_Holiday = dframe[dframe['Holiday_Flag'] == 1].groupby('Store')['Weekly_Sales'].sum().astype(int)
Avg_NormalWeek_sale = dframe[dframe['Holiday_Flag']==0].groupby('Store')['Weekly_Sales'].mean().astype(int)
Avg_HolidayWeek_sale = dframe[dframe['Holiday_Flag'] == 1].groupby('Store')['Weekly_Sales'].mean().astype(int)

stores = Avg_NormalWeek_sale.index #storing id 
plt.figure(figsize=(10,7))
bar_width = 0.4
x = np.arange(len(stores))

plt.bar(x - bar_width / 2, Avg_NormalWeek_sale, width=bar_width, label='Average Sales on a normal week ',color='green')
plt.bar(x + bar_width / 2, Avg_HolidayWeek_sale, width=bar_width, label='Average Sales from just holiday week',color='orange')
plt.xlabel('Stores')
plt.ylabel('Average Weekly Sales')
plt.title('Average Weekly Sales: Holiday vs Non-Holiday Weeks')
plt.xticks(x, stores, rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
